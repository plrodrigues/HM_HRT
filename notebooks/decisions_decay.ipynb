```python
initial_probability = 0.9
minimum_probability = 0.05
decay_factor = 0.95


def exp_func(x):
    return max(initial_probability * (decay_factor**x), minimum_probability)


exp_result = [exp_func(x) for x in range(0, 60)]
```

```python
plt.plot(exp_result)
```
