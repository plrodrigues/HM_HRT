# Decide what decay function to use on the probability for the mutation

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import jupyter_black

jupyter_black.load()
```

```python
initial_probability = 0.9
minimum_probability = 0.05
decay_factor = 0.95


def exp_func(x):
    return max(initial_probability * (decay_factor**x), minimum_probability)


exp_result = [exp_func(x) for x in range(0, 60)]
```

```python
plt.figure()
plt.plot(exp_result, color="b")
plt.xlabel("n interation")
plt.ylabel("probability")
plt.title(f"Decay of the probability per iteration")
plt.text(3, 0.85, "y = max(initial_prob. * (decay_factor**x), min_prob.)", fontsize=12)
plt.show()
```
